```python
# Projeto 1 - Desenvolvimento de Game em Linguagem Python

# Import
import random
from os import system, name
```

```python
# Função para limar a tela a cada execução
def limpa_tela():
 
    # Windows
    if name == 'nt':
        _ = system('cls')
 
    # Mac ou Linux
    else:
        _ = system('clear')
```

```python
# Função que desenha a forca na tela
def display_hangman(chances):

    # Lista de estágios da forca
    stages = [  # estágio 6 (final)
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """,
                # estágio 5
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                """,
                # estágio 4
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                """,
                # estágio 3
                """
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                """,
                # estágio 2
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                # estágio 1
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                # estágio 0
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return stages[chances]        
```

```python
# Função jogo
def game():

    limpa_tela()
    print("\nBem-vindo(a) ao jogo da forca!")
    print("Adivinhe a palavra abaixo:\n")

    # Lista de palavras para o jogo
    palavras = ['abacaxi','melancia','manga','abacate', 'uva','goiaba','morango', 'laranja']
    
    # Escolhe randomicamente uma palavra
    palavra = random.choice(palavras)

    # Preencher dos vazios conforme a quantidade de letras da palavra
    letras_descobertas = ['_' for letra in palavra]

    # Número de chances
    chances = 6

    # Lista para as letras erradas
    letras_erradas = []

    # Loop enquanto número de chances for maior do que zero
    while chances > 0:

        # Imprimindo as informações para o usuário
        print(display_hangman(chances))
        print(" ".join(letras_descobertas))
        print("\nLetras erradas:", " ".join(letras_erradas))
        print("\nChances restantes:", chances)
        

        # Tentativa do usuário
        tentativa = input("\nDigite uma letra: ").lower()

        # Condicional
        
        #se a letra digitada estiver contida na palavra:
        if tentativa in palavra:
            
            #zere o contador
            index = 0
            
            #para cada letra contida na palavra:
            for letra in palavra:
                
                #se a letra escolhida for igual a letra selecionada:
                if tentativa == letra:
                    
                    #substitua o espaço vazio na posição atual pela letra escolhida
                    letras_descobertas[index] = letra
                    
                #incremento    
                index += 1
        else:
            #Decremento
            chances -= 1
            
            #adicione a letra escolhida a lista de letras erradas
            letras_erradas.append(tentativa)

        # Condicional
        
        #se não houver mais espaços vazios, imprima que o usuário venceu o jogo
        if "_" not in letras_descobertas:
            print("\nVocê venceu, a palavra era:", palavra)
            break

    # Condicional
    #caso acabe o numero de chances e ainda existam espaços vazios, imprima que o usuário perdeu o jogo
    if "_" in letras_descobertas:
        print("\nVocê perdeu, a palavra era:", palavra)
```

```python
# Bloco main
if __name__ == "__main__":
    game()
```
